-------------------------
### Import Library
------------------------

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np
from tensorflow.keras import Sequential
from tensorflow import keras
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

------------------------
### Load Dataset dan Split Data
----------------------

```python
(x_train_raw, y_train_raw),(x_test_raw, y_test_raw) = tfds.load("uc_merced", batch_size = -1, 
                                                                split=['train', 'train[:70%]'],
                                                               as_supervised=True)
```

```python
numclass = 21

# One Hot Encoding
y_train = keras.utils.to_categorical(y_train_raw, numclass)
y_test = keras.utils.to_categorical(y_test_raw, numclass)
```

-------------------
### Preview Gambar
--------------------

```python
for i in range(20):
    plt.subplot(5,4,i+1)
    plt.imshow(x_train_raw[i])
    plt.axis('off')
plt.show()
```

```python
x_train_raw.shape
```

```python
x_test_raw.shape
```

```python
train_data = tf.image.resize(x_train_raw, (28, 28))
test_data = tf.image.resize(x_test_raw, (28, 28))
```

```python
train_data = np.reshape(train_data.numpy(), (2100, 2352))
test_data = np.reshape(test_data.numpy(), (1470, 2352))
```

```python
# Normalisasi piksel pada gambar
train_data = train_data.astype('float32')/255
test_data = test_data.astype('float32')/255
```

------------------------
### Membuat DNN ( Deep Neural Network )
---------------------------

```python
modeldata = Sequential([
        Dense(512, activation = 'relu', input_dim = (2352)),
        Dense(512, activation = 'relu'),
        Dense(numclass, activation = 'softmax')
    ])
```

```python
# Menunjukan kesimpulan dari model
modeldata.summary()
```

-------------------
### Optimisasi Model
----------------

```python
modeldata.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics=['accuracy'])
```

--------------------
### Fitting Training Model
------------------------

```python
data_fit = modeldata.fit(
    train_data, 
    y_train, 
    batch_size = 512, 
    epochs = 20, 
    validation_data = (test_data, y_test)
)
```

---------------------------------------
### Evaluasi Model Yang Telah Di Train
-----------------------------------------------

```python
score = modeldata.evaluate (test_data, y_test)
print("TEST LOSS     :", score[0])
print("TEST ACCURACY :", score[1])
```
